## Bitwise Operators

- These operators manipulate individual bits of numbers.
- Integers are converted into binary and then these operations are performed on individual bits.

> **Note:** Convert all Decimal Numbers to 8-bit Binary Representations

| Operator | Meaning |
|----------|---------|
| & | Bitwise AND |
| \| | Bitwise OR |
| ^ | Bitwise exclusive OR / Bitwise XOR |
| ~ | Bitwise inversion (one's complement) |
| << | Shifts the bits to left / Bitwise Left Shift |
| >> | Shifts the bits to right / Bitwise Right Shift |

````python
# Bitwise AND, OR and XOR
a = 7
b = 3

# Try working out the solutions manually first.
print(a & b)
print(a | b)
print(a ^ b)

# Bitwise Left Shift
c = 10
print(c << 3)
print(c >> 1)
````

````python
print(~c)
````

### Bitwise NOT - Why ~10 = -11?

Simple Forumula:
```
~n = -(n + 1)
```

To get binary representation of a number:
```python
num = 5
print(bin(num))
```

Computer's process to read 11110101:
1. "This starts with 1, so it's negative"
2. "To find the value, I'll flip all bits: 11110101 → 00001010"
3. "Then add 1: 00001010 + 1 = 00001011"
4. "00001011 in decimal is 11"
5. "Since original started with 1, the answer is -11"

### Basic Syntax/Skeleton

```text

```

````python
# IF Statement
balance = int(input("Enter your balance: "))

if balance < 100:
    print("Broke 🥲")
    
print("Done")
````

````python
# If Else Statement
RCB_WINS = True

if RCB_WINS:
    print("Ee saala cup namde")
else:
    print("Boys played well")
````

````python
# IF - ELSE IF - ELSE Ladder
marks = int(input("Enter marks: "))

if marks >= 75:
    print("Aal izz well!")
elif marks >= 35:
    print("Pass hogaye")
else:
    print("Abba nahi mannenge")
````

````python
# Nested IF

username = "password" 
password = "username" 

if username == 'password':
    if password == "username":
        print("Logged in")
    else:
        print("Wrong password")
else:
    print("Username not found")
````

### Python Comparison Operators

| Operator | Name | Description | Example | Result |

| `==` | Equal to | Checks if values are equal | `5 == 5` | `True` |

## Ternary Operator

- How it is in C++
  ```text
  variable = (condition) ? True Statement : False Statement
  ```

  **Example:**
  ```cpp
  int num;
  bool even = (num % 2 == 0) ? True : False;
  ```
---
- We do have the same functionality in Python:

  **Syntax:**
  ```text
  variable = value_1 if condition else value 2
  ```
---

````python
# Easy Way -> So many lines 😵
res = ''
num = 10
if num % 2 == 0:
    res = "Yes"
else:
    res = "No"

res
````

````python
# Efficient Way - Single Line
res = "Yes" if num % 2 == 0 else "No"
print(res)
````

## For Loop
**Syntax:**
```text
for variable in iterable:
    // statements
```

**Q**: What is `Iterable` now?? <br>
**A**: Object in which we you can go through the elements one by one (list, tuple, set, array, string).

````python
# Sensors Example
sensors = ['DHT11', 'IR-Sensor', 'Soil-Moisture-Sensor', 'pH-Sensor', 'Ultrasonic-Sensor', 'NPK-Sensor']

for sensor in sensors:
    print(sensor)
````

````python
print('List of all sensors:')
for sensor in sensors:
    print(sensor, end=' ')
````

````python
# Iterating over String
name = "Jethalal"

for ch in name:
    print(ch, end = ' ')
````

## range() Function

- ```range()``` function generates sequence of numbers.
- Syntax:
  1. ```range(stop)```: 0 to stop - 1
  2. ```range(start, stop)```: start to stop - 1
  3. ```range(start, stop, step)```: start to stop - 1 with step

---

### Looping with Range
- How we did in C/C++:
  ```c++
  for(int i = 0; i < 10; i++){
      cout<<i<<endl;
  }
  ```
<br>

- How we do it in Python:
    ```python
    for i in range(10):
        print(i)
    ```

````python
for i in range(10):
    print(i, end=' ')
````
