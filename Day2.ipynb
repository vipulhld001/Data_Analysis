````python
## Hangman

import random
from hangman_art import stages, logo # Import the Py Files for Stages and Logos
from hangman_words import word_list # Import py file for word list


print(logo) #Show the logo of the game


````

````python
game_is_finished = False

lives = len(stages) - 1 # Use stages to show that we have the guy hanging emoticon

chosen_word = random.choice(word_list) # Select the word at random
word_length = len(chosen_word)         # Find the length to get the ___ Values

print(stages[lives])
````

````python
display = []
for _ in range(word_length):
    display += "_"
````

````python
while not game_is_finished:
    guess = input("Guess a letter: ").lower()

    if guess in display:
        print(f"You've already guessed {guess}")

    for position in range(word_length):
        letter = chosen_word[position]
        if letter == guess:
            display[position] = letter
    print(f"{' '.join(display)}")

    if guess not in chosen_word:
        print(f"You guessed {guess}, that's not in the word. You lose a life.")
        lives -= 1
        if lives == 0:
            game_is_finished = True
            print("You lose.")
    
    if not "_" in display:
        game_is_finished = True
        print("You win.")
````

# Arbitrary Arguments - *args
## If you do not know how many arguments will be passed into your function, add a * before the parameter name.

## This way, the function will receive a tuple of arguments and can access the items accordingly:

````python
def my_function(*kids):
  print("The youngest child is " + kids[1])

my_function("Emil", "Tobias", "Linus")
````

# Use `*args` with another argument

````python
def my_function(greeting, *names):
  for name in names:
    print(greeting, name)

my_function("Hello", "Emil", "Tobias", "Linus")
````

# WAP to Calculate the sum on any number of values

````python
def add_all(*numbers):
  total = 0
  for num in numbers:
    total += num
  return total

print(add_all(1, 2, 3))
print(add_all(10, 20, 30, 40))
print(add_all(5))
````

# List Comprehension

List comprehension offers a shorter syntax when you want to create a new list based on the values of an existing list.

Example:

Based on a list of fruits, you want a new list containing only the fruits with the letter "a" in the name.

Without list comprehension, you will have to write a `for` statement with a conditional test inside:

````python
fruits = ["apple", "banana", "cherry", "kiwi", "mango"]
newlist = []

for x in fruits:
  if "a" in x:
    newlist.append(x)

print(newlist)
````

````python
newlist = [x for x in fruits if "a" in x]
````

# **Mutability**

# Immutable objects can **NOT** be changed after creation

````python
short_listed_students = ("Ajay", "Aman", "Puja", "Sakshi")
print(short_listed_students)
type(short_listed_students)

short_listed_students = ("Ajay","Ajay", "Aman", "Puja", "Sakshi")
print(short_listed_students)
type(short_listed_students)
````

````python
thistuple = ("apple",)
print(type(thistuple))

#NOT a tuple
thistuple = ("apple")
print(type(thistuple))
````
