# **Loc vs iloc**
# What is .loc[]?
 - # Used for label-based data selection
 - # We select data by row or column names

```python
# loc[] structure is as (df.loc[rwo_label, column_label])
df.loc[2,['Country','Age', 'LanguageWorkedWith']]
#df.columns
```

# Why we use .loc[]?
 - # Easy to use when working with named indices or column headers
 - # Supports label slicing and conditional access

```python
df.loc[df['Age']>30]
```

# What is .iloc[]?
 - # Used for position-based selection
 - # We select data by integer position (like list indexes)

```python
# structre of .iloc[] (df.iloc[row_index, column_index])
df.iloc[0,1] # This shows the first row, second column value
```

# Why we use .iloc[]?
 - # Precise and consistent with zero-based indexing
 - # Useful when labels are unknown or sequential

```python
df.iloc[0:15] # showe the first 3 rows
```

```python
df.loc[[0]] # show first row
df.loc[[0,2]] # first row and third row
df.loc[[0,2], 'Country'] # [[r,r],c]
df.loc[1:10, 'MainBranch':'Hobbyist']
```

# **Changing index**


```python
students = {
    "S_name":["Akash","Puja","Lina","Rawina"],
    "S_ID":["2026cs001","2026cs002","2026cs003","2026cs004"],
    "S_email":["akash.in@gmail.com","puja@gmail.com","lina@gmail.com","rawina@gmail.com"]}
sdf = pd.DataFrame(students)
```

```python
sdf.index # to check the index in the sdf dataframe
```

# **Point** Make sure that the column you want to change to an index has unique values

```python
sdf.set_index('S_ID') # this will change the index, but it won't be permanently applied
```

```python
sdf.set_index('S_ID', inplace=True) # using inplace = True the assigned column will be permenantly converted to index
```

```python
sdf.loc['2026cs001'] # This will retrieve the record for the student with the given id
```

```python
sdf.iloc[0] # index based retrival while .loc[] uses label based retrival
```

```python
sdf.reset_index(inplace=True) # this will reset the index to its original form
```
